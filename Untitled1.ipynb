```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing, metrics
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from keras.models import Sequential
from keras.layers import Dense, Input, CategoryEncoding, Dropout
from keras.utils import to_categorical
from keras import regularizers
import keras
from xgboost import XGBClassifier, plot_importance
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
df = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

```python
df.info()
```

```python
df_nn = df.copy()
```

```python
df_nn.info()
```

```python
cat_cols = []
num_cols= []
for i in df_nn.columns:
    if i != 'Gender':
        if df_nn[i].dtype == 'object':
            cat_cols.append(i)
        else:
            num_cols.append(i)
```

```python
num_cols
```

```python
cat_cols
```

```python
corr_matrix = df_nn.corr(numeric_only=True)
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Matriz de Correlación - Variables Numéricas")
plt.show()
```

```python
#df_nn.drop(['NumCompaniesWorked'], axis = 1, inplace = True)
```

```python
df_encoded = pd.get_dummies(df_nn, columns=cat_cols, drop_first=True)
```

```python
df_encoded
```

```python
df_encoded.info()
```

```python
y_df = df_encoded['Attrition_Yes']
x = df_encoded.drop(['Attrition_Yes', 'Gender'], axis = 1)
y = y_df.values
```

```python
y.shape
```

```python
x.shape
```

```python
y_df
```

```python
y
```

```python
no_categories = len(y_df.unique())
```

```python
label_encoder = preprocessing.LabelEncoder()
y_df_encoded = label_encoder.fit_transform(y_df)
```

```python
scaler = preprocessing.MinMaxScaler()
x_scaled = scaler.fit_transform(x)
```

```python
x_train, x_val_test, y_train, y_val_test = train_test_split(
    x_scaled, y_df_encoded, test_size=0.3, random_state=42, stratify=y_df_encoded)
```

```python
x_val, x_test, y_val, y_test = train_test_split(
    x_val_test, y_val_test, test_size=0.5, random_state=42, stratify=y_val_test)
```

```python
y_train_categorical = to_categorical(y_train, num_classes=no_categories)
y_val_categorical = to_categorical(y_val, num_classes=no_categories)
y_test_categorical = to_categorical(y_test, num_classes=no_categories)
```

```python
y_train.shape
```

```python
no_categories
```

```python
model = Sequential([
    Input(shape=(x_train.shape[1],)),
    Dense(128, activation='relu', kernel_regularizer = regularizers.l1(0.001)),
    Dropout(0.3),
    Dense(128, activation='relu', kernel_regularizer = regularizers.l1(0.001)),
    Dropout(0.3),
    Dense(128, activation='relu', kernel_regularizer = regularizers.l1(0.001)),
    Dropout(0.2),
    Dense(no_categories, activation='sigmoid')
])
```

```python
model.compile(optimizer='adam', 
              loss='binary_crossentropy', 
              metrics=['accuracy'])
```

```python
hist = model.fit(x_train, y_train_categorical,
                batch_size = 16, epochs = 50,
                validation_data = (x_val, y_val_categorical))
```

```python
model.evaluate(x_test, y_test_categorical)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred_cat = np.argmax(y_pred, axis = 1)
```

```python
y_pred_cat.shape
```

```python
metrics.confusion_matrix(y_test, y_pred_cat)
```

```python
sns.heatmap(metrics.confusion_matrix(y_test, y_pred_cat), annot = True, cmap = 'viridis')
plt.show()
```

```python
k_values = range(1,21)
accuracies = []

for k in k_values:
    knn_temp = KNeighborsClassifier(n_neighbors = k)
    knn_temp.fit(x_train, y_train_categorical[:,1])
    y_pred_temp = knn_temp.predict(x_test)
    accuracies.append(accuracy_score(y_test, y_pred_temp))

plt.figure(figsize = (10,6))
plt.plot(k_values, accuracies, marker = 'o', linestyle = '-')
plt.xticks(k_values)
plt.ylim((0,1))
plt.grid()
plt.show()
```
